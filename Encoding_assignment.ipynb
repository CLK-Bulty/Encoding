```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
df2 = df.copy()
df3 = df.copy()
df4 = df.copy()
```

#Manual encoding

##Label encoding

```python
categorical_cols = ['sex', 'smoker', 'region'] 
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df[categorical_cols] = df[categorical_cols].apply(lambda col: 
le.fit_transform(col)) 
```

```python
df.head()
```

#One hot encoding

```python
categorical_cols = ['sex', 'smoker', 'region'] 
```

#Ordinal encoding


```python
df3.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
genders = df3.sex.unique()
```

```python
genders
```

```python
ordinal = OrdinalEncoder(categories=[genders])
```

```python
ordinal
```

```python
encoded = ordinal.fit_transform(df3[['sex']])
```

```python
encoded
```

```python
df3['sex'] = pd.DataFrame(encoded, columns=['sex'])
```

```python
df3.head()
```

```python
smok = df3.smoker.unique()
```

```python
ordinal = OrdinalEncoder(categories=[smok])
```

```python
ordinal
```

```python
encoded = ordinal.fit_transform(df3[['smoker']])
```

```python
encoded
```

```python
df3['smoker'] = pd.DataFrame(encoded, columns=['smoker'])
df3.head()
```

```python
reg = df3.region.unique()
```

```python
ordinal = OrdinalEncoder(categories=[reg])
```

```python
ordinal
```

```python
encoded = ordinal.fit_transform(df3[['region']])
```

```python
df3['region'] = pd.DataFrame(encoded, columns=['region'])
df3.head()
```

#Encoding using for loop

#Label encoding

```python
df4.head()
```

```python
label_encoders = {}
categorical_columns = ['sex', 'smoker', 'region']  

for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    df4[column] = label_encoders[column].fit_transform(df4[column])
```
